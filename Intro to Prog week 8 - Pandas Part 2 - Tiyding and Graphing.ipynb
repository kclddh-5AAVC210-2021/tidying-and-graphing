```python
# Import pandas, numpy and seaborn

import pandas as pd
import numpy as np
import seaborn as sns

# Use seaborn's default graphing theme
sns.set_theme()
```

```python
# A very simple dataframe

# Now let's try creating a two-dimenstional dataframe

people = {
    'name' : ["Anna", "Baz", "Claire", "Daniel", "Edwin", "Fred", "Georgia", "Harry", "India", 'Jackie'],
    'gender'  : ["female", "nonbinary", "female", np.nan, "intersex", np.nan, "male", "nonbinary", "male", "female"],
    'age'  : [np.nan, 28, np.nan, 55, 18, 21, 26, 62, 12, 34],
    'height' : [174.21, 156.41, np.nan, 170.14, 159.64, 163.21, 201.56, 177.51, 178.41, 165.01],
    'Where they live': ["Lagos", "Phnom Penh", "Seoul", np.nan, "London", "Vancouver", "Scarborough", "Chicago", "London", "Manchester"],
    'grade'  : ["B", "D", "A*", "A", "C", "B", "A", "B", "A", "E"],
}

friends = pd.DataFrame(people)

friends
```

```python
# Recap! 

# Filter just certain columns

friends[["name", "age", "grade"]]
```

```python
friends
```

```python
# Recap! 

# Find a certain friend

friends[friends["name"] == "Georgia"]
```

```python
# Group by

friends.groupby("gender").count()
```

```python
# Drop all rows with NaN's in them
friends.dropna()
```

```python
# Drop rows with NaNs in the height column 
friends.dropna(subset=['height']) 
```

```python
# Backfill NaN's with the values that came before them
# See https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html

friends.fillna(method="bfill")
```

```python
# Backfill na's with the values that came before them

friends.fillna(method="ffill")
```

```python
friends['gender'] = friends['gender'].replace(np.nan,'Unknown')
```

# Data visualisation time! 

As we've already loaded up seaborn as sns in our first cell above, let's go straight to opening up a dataframe. This time the built in penguins dataframe

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.histplot(
    data=penguins, 
    x="species",
    hue="island",
    multiple="stack"
)
```

```python
# Let's draw a histogram of all the penguins flipper lengths, 
# divided by ther species

sns.histplot(
    data=penguins, 
    x="body_mass_g",
    hue="species",
    multiple="stack"
)
```

```python
# We can break this up into multiple graphs with
# displot() and col="<value>"

sns.displot(
    data=penguins, 
    x="island", 
    col="species",
    hue="island"
)
```

```python
# Let's draw a scatter diagram
sns.relplot (
    data=penguins,
    kind="scatter",
    x="flipper_length_mm",
    y="body_mass_g",
    hue="sex",
    style="species",
    size="body_mass_g"
)
```

```python
# How about a line diagram?

sns.relplot (
      data=penguins,         # set the dataframe
      kind="line",           # type of graph. Alternatives: scatter
      x="flipper_length_mm", # set the x coordinate
      y="body_mass_g",       # Set the y coordinate
      hue="island"
)
```

```python
# 

sns.displot(
    data=penguins, 
    x="flipper_length_mm",
    hue="sex",
    multiple="stack",
    kde=True
)
```

```python
sns.catplot(
    data=penguins,    
    kind="swarm", 
    x="island",
    y="body_mass_g",
    hue="species",
)
```

```python
sns.catplot(
    data=penguins, 
    kind="violin", 
    x="island",
    y="flipper_length_mm",
    hue="species",
)
```

```python
sns.catplot(
    data=penguins, 
    kind="bar", 
    x="species",
    y="flipper_length_mm",
)
```

```python
sns.jointplot(
    data=penguins,
    x="flipper_length_mm",
    y="bill_length_mm",
    hue="species"
)
```

```python
sns.pairplot(data=penguins,
             hue="species")
```
